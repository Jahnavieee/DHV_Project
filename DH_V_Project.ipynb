```python
# Droping unecessary columns 
ds = ds.drop(columns =  ['in_apple_playlists','in_apple_charts','in_deezer_playlists','in_deezer_playlists','in_shazam_charts','in_deezer_charts'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table
key_columns = ['artist_count', 'released_year', 'released_month', 'released_date', 'in_spotify_playlists', 'in_spotify_charts', 'bpm'
              , 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%' ]
statistics = ds[key_columns].describe()

# Generate summary statistics
#statistics = df.describe()

# Create a figure
fig, ax = plt.subplots(figsize=(10, 6), dpi=200)

# Hide axes
ax.axis('off')

# Create the table
stats_table = table(ax, statistics, loc='center', colWidths=[0.2] * len(statistics.columns))

# Formatting the table
stats_table.auto_set_font_size(False)
stats_table.set_fontsize(10)  # Set font size
stats_table.scale(1.2, 1.2)  # Scale the table for better readability

# Add title
plt.title("Summary Statistics", fontsize=14, fontweight='bold')

# Save as an image
plt.savefig('summary_statistics.png', bbox_inches="tight", dpi=300)

# Display the image
plt.show()
```

```python
# Plot 1: Distribution of KEYS across other playlist
plt.figure(figsize=(8, 6))
sns.histplot(ds['key'], kde=True, color='red', bins=20)
plt.title("Distribution of Keys")
plt.xlabel("Key")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot 2: Average released date of songs by the artist 
plt.figure(figsize=(8, 6))
sns.barplot(data=ds, x='artist_count', y='released_date', palette='rocket_r')
plt.title("Average Artist Count By Their Released Date")
plt.xlabel("Artist Count")
plt.ylabel("Average Released Date")
plt.show()
```

```python
# Plot 3: Heatmap of correlations among numerical features
plt.figure(figsize=(12, 8))

numerical_features = ds.select_dtypes(include=np.number)

correlation_matrix = numerical_features.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="cubehelix", cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Plot 5: Box Plot of modes by the playlist
plt.figure(figsize=(8, 6))
sns.boxplot(data=ds, x='mode', y='in_spotify_playlists', palette='mako')
plt.title("Modes in Spotify Playlist")
plt.xlabel("Modes")
plt.ylabel("Spotify Playlist")
plt.show()
```

```python
# Plot 6:Scatter Plot
sns.set()
fig, axes = plt.subplots(1,4, figsize=(18,5))
fig2, axes2 = plt.subplots(1,3, figsize=(11.4,5))
plt.setp(axes, xlim=(-5, 105))
plt.setp(axes2, xlim=(-5, 105))
fig.suptitle('Song characteristics against spotify playlist')
song_characteristics = ['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
for i in range(1):
    for j in range(4):
        sns.scatterplot(x=song_characteristics[i*4 + j],
                y='in_spotify_playlists', data=ds, ax=axes[j], color= 'purple')
for i in range(1,2):
    for j in range(3):
        sns.scatterplot(x=song_characteristics[i*4 + j],
                y='in_spotify_playlists', data=ds, ax=axes2[j], color='purple')
```
